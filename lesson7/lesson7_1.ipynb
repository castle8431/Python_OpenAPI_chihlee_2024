```python
import requests #https://requests.readthedocs.io/en/stable/
from requests import Response
from requests import RequestException #https://requests.readthedocs.io/en/stable/user/quickstart/#errors-and-exceptions
url = 'https://data.ntpc.gov.tw/api/datasets/75abd5e2-5454-4c59-973f-ba587ef94ccd/csv?page=0&size=1000'
#http request
#GET 資料在網址?後面 明碼 危險
#post 資料打包 資料傳輸攔截一樣會被抓到 需要做加密動作:打包變成看不懂的東西 必須解密要有key, key只有傳輸給對方的server有
#http沒加密 https有加密
#http response回傳 (回傳完就斷線)
try:
    r:Response = requests.request('GET',url)
    if r.status_code == 200:
        print("下載成功")
    else:
        print("下載有問題")
except RequestException as e:
    print(e)
```

```python
#改用raise_for_status https://requests.readthedocs.io/en/stable/user/quickstart/#errors-and-exceptions
import requests 
from requests import Response
from requests import RequestException, HTTPError 
url = 'https://data.ntpc.gov.tw/api/datasets/75abd5e2-5454-4c59-973f-ba587ef94ccd/csv?page=0&size=1000'

try:
    r:Response = requests.request('GET',url)
    r.raise_for_status()
except HTTPError as e:
    print(e)
except RequestException as e:
    print(e)
else:
    print("下載成功")
```

```python
print(type(r.text))
```

#python的資料結構為list, dict, set

```python
from io import StringIO #https://docs.python.org/zh-tw/3.13/library/io.html
file = StringIO(r.text)
file
```

```python
from csv import DictReader #https://docs.python.org/zh-tw/3.13/library/csv.html
reader = DictReader(file) #reader:Generator實體 被讀過一次之後記憶體就消失
reader
```

```python
for row in reader:
    print(type(row)) #這邊reader被讀取過記憶體就沒了
```

```python
from csv import DictReader #https://docs.python.org/zh-tw/3.13/library/csv.html
reader = DictReader(file) 
list_reader:list[dict] = list(reader) #list記憶體資料永遠保留
```

```python
for row in list_reader:
    print(row)
```
